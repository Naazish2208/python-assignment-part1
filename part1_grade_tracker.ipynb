```python
# Task 1: Data Parsing & Profile Cleaning

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
# Cleaning the inconsistencies

cleaned_students = []
for student in raw_students:
  cleaned_rows = {
      "name": student["name"].strip().title(),
      "roll": int(student["roll"]),
      "marks": [int(mark) for mark in student["marks_str"].split(", ")]
  }

  cleaned_students.append(cleaned_rows)
  print(cleaned_rows)
```

```python
# Validating the students' names

for student in cleaned_students:
  is_valid = all(char.isalpha() for char in student["name"].split())
  validity = "✓ Valid name" if is_valid else "✗ Invalid name"
  print(f"{student['name']}: {validity}")
```

```python
# Creating formatted profile card for each student

for temp in cleaned_students:
 print(f"================================")
 print(f"Student: {temp['name']}")
 print(f"Roll No: {temp['roll']}")
 print(f"Marks: {temp['marks']}")
 print(f"================================\n")
```

```python
# Printing Roll no. 103 in UPPER and lower case

for student in cleaned_students:
  if student["roll"] == 103:
    print(f"Roll No. 103 in UPPER CASE: {student['name'].upper()}")
    print(f"Roll No. 103 in lower CASE: {student['name'].lower()}")
```

```python
# Task 2: Marks Analysis Using Loops & Conditionals

student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]
```

```python
# Assigning grade labels to marks

for i in range(len(subjects)):
  subject = subjects[i]
  mark = marks[i]

  if mark >= 90:
    grade = "A+"
  elif mark >= 80:
    grade = "A"
  elif mark >= 70:
    grade = "B"
  elif mark >= 60:
    grade = "C"
  else:
    grade = "F"
  print(f"{subject}: {mark}, Grade: {grade}")
```

```python
# Calculating and printing

total_marks = sum(marks)
print(f"Total Marks: {total_marks}")

average_marks = round(total_marks / len(subjects), 2)
print(f"Average Marks: {average_marks}")

highest_mark = max(marks)
highest_sub = subjects[marks.index(highest_mark)]
print(f"Highest Scoring Subject: {highest_sub} - {highest_mark}")

lowest_mark = min(marks)
lowest_sub = subjects[marks.index(lowest_mark)]
print(f"Lowest Scoring Subject: {lowest_sub} - {lowest_mark}")
```

```python
# Simulating marks-entry system

new_subject_count = 0

while True:
  new_subject = input("Enter the name of the new subject (or type 'done' to end the process): ")

  if new_subject.lower() == 'done':
    break

  try:
    new_mark = int(input(f"Enter the mark for {new_subject} (0-100): "))
    if 0 <= new_mark <= 100:
      subjects.append(new_subject)
      marks.append(new_mark)
      new_subject_count += 1
    else:
      print("Do not crash, and do not add invalid entries to the list")
  except ValueError:
    print("Do not crash, and do not add invalid entries to the list")

print(f"{new_subject_count} new subject(s) added.")

updated_average = round(sum(marks) / len(subjects), 2)
print(f"Updated Average Marks: {updated_average}")
```

```python
# Task 3: Class Performance Summary

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
# Assign Pass or Fail

summary_table = []

print(f"{'Name':<18} | {'Average':^7} | {'Status'}")
print("-" * 40)

for Name, marks_list in class_data:
  Average = round(sum(marks_list) / len(marks_list), 2)
  Status = 'Pass' if Average >= 60 else "Fail"
  summary_table.append((Name, Average, Status))
  print(f"{Name:<18} | {Average:^7.2f} | {Status}")


```

```python
# Students who passed, failed, topped and the class average

pass_count = 0
fail_count = 0
total_average = 0

for student in summary_table:
  if student[2] == "Pass":
    pass_count += 1
  else:
    fail_count += 1
  average = round(student[1], 2)
  total_average += average

class_topper = max(summary_table, key=lambda x: x[1])

print(f"Number of students who passed: {pass_count}")
print(f"\nNumber of students who failed: {fail_count}")
print(f"\nClass Topper: {class_topper[0]} - {class_topper[1]}")
print(f"\nClass Average: {total_average}")
```
